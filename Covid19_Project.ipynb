```python
# Check unique countries 
countries = list(covid_dataset['Country/Region'].unique())
print(countries)
# length of countries
print("Number of Countries: ", len(countries))
```

```python
aggregation = covid_dataset.groupby('Country/Region').sum()

# Display the shape of the aggregated DataFrame
print(aggregation.shape)

# Display the aggregated DataFrame
print(aggregation)
```

```python
selected_date = '4/21/20'

# Create a bar plot
sns.barplot(x='Country/Region', y=selected_date, data=covid_dataset)
plt.xticks(rotation=90)
plt.xlabel('Country/Region')
plt.ylabel(f'Number of Cases on {selected_date}')
plt.title(f'COVID-19 Cases on {selected_date} by Country/Region')
plt.show()
```

```python
sns.relplot(x="Country/Region" , y="4/21/20" ,  data= covid_dataset )
```

```python
# Plot for the first n days
n = 3
aggregation.loc['India'][:n].plot()
```

```python
country_name = 'India'
n = 3

# Plot the last n days
aggregation.loc[country_name][-n:].plot()
plt.xlabel('Date')
plt.ylabel(f'Number of Cases ({country_name})')
plt.title(f'COVID-19 Cases in {country_name} for the Last {n} Days')
plt.show()
```

```python
country_name = 'India'
start_date = '3/16/20'
end_date = '3/26/20'

# Plot for the specified date range
aggregation.loc[country_name, start_date:end_date].plot()
plt.xlabel('Date')
plt.ylabel(f'Number of Cases ({country_name})')
plt.title(f'COVID-19 Cases in {country_name} from {start_date} to {end_date}')
plt.show()
```

```python
# Find the data for a particular country and plot it
country_name = 'India'
country_data = covid_dataset[covid_dataset['Country/Region'] == country_name]

sns.relplot(x='Country/Region', y='4/21/20', data=country_data, kind='scatter')
plt.xlabel('Country/Region')
plt.ylabel('Number of Cases on 4/21/20')
plt.title(f'COVID-19 Cases on 4/21/20 for {country_name}')
plt.show()
```

```python
# Calculate the maximum and minimum increase and average in the cases
max_inc = aggregation.loc['India'].diff().max()
min_inc = aggregation.loc['India'].diff().min()
avg_inc = aggregation.loc['India'].diff().mean()

print("Maximum Increase in a day:", max_inc)
print("Minimum Increase in a day: ", min_inc)
print("Average Increase in a day: ", avg_inc)
```
